```python
import pandas as pd
```

```python
# create dataframe using dictionary

store_data = pd.DataFrame({'CustomerId': ['CustId01', 'CustId02', 'CustId03', 'CustId04', 'CustId05'],
                           'location': ['Chicago', 'Boston', 'Seattle', 'San Franscisco', 'New York'],
                           'gender': ['M', 'M', 'F', 'M', 'F'],
                           'type': ['Electronics', 'Food & Beverages', 'Food & Beverages', 'Medicine', 'Beauty'],
                           'quantity': [1,3,4,2,1],
                           'total_bill': [100, 75, 125, 50, 80]})
store_data
```

```python
# 1st row of Dataframe
store_data.iloc[:1]
```

```python
# 1st column of Dataframe
store_data['location']
```

```python
# accessing rows with the step size of 2
store_data.iloc[::2]
```

```python
# accessing rows in reverse order
store_data.iloc[::-1]
```

```python
# loc method
# accessing 1st index value using loc method (index starts from 0)
store_data.loc[1]
```

```python
# accessing 1st, 4th index values along with location and type columns

store_data.loc[[1,4], ['location', 'type']]
```

```python
store_data
```

```python
#iloc
'''
used for integer location-based inexing/selection by position.
When using the loc method on a dataframe, we specify which rows and columns we want by using the following format:
    dataframe.iloc[row selection, column selection]

    loc - label based
    iloc - integer position value based (0-based integer position)
'''
store_data.iloc[[1,4],[0,2]]
```

```python
store_data.loc[4, 'location']
```

```python
store_data
```

```python
print(store_data.loc[4 , 'type'])
```

```python
# modify entries using loc

store_data.loc[4, 'type'] = 'Electronics'
store_data
```

```python
# modify entries using iloc

store_data.iloc[4,3] = 'Beauty'
store_data
```

```python
# condition based indexing

store_data['quantity'] > 1
```

```python

store_data.loc[store_data['quantity'] > 1]
```

```python
store_data
```

```python
# adding a new column in dataframe

store_data['rating'] = [2,5,3,4,4]
store_data
```

```python
# removing a new column in dataframe
# axis=1, drops the column
# axis=0, drops the row
store_data.drop('rating', axis=1)
store_data
```

```python

# drop the column permanently from the dataframe. use, inplace=True
store_data.drop('rating', axis=1, inplace=True)
store_data
```

```python
new_store_data = store_data.copy()
new_store_data
```

```python
# dropping multiple columns

new_store_data.drop(['location', 'quantity'], axis=1, inplace=True)
new_store_data
```

```python
store_data
```

```python
new_store_data_2 = store_data.drop(1, axis=0)
new_store_data_2
```

```python
# resetting the index of the dataframe (a new column 'index' got added)
new_store_data_2.reset_index()
```

```python
# drop the newly created column 'index'

new_store_data_2.reset_index(drop=True, inplace=True)
new_store_data_2
```
